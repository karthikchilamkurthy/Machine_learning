```python
i_s = iris.loc[iris['variety']=='Setosa']
i_vi = iris.loc[iris['variety']=='Virginica']
i_ve = iris.loc[iris['variety']=='Versicolor']
```

```python
plt.plot(i_s['petal.length'],np.zeros_like(i_s['petal.length']),'o')
plt.plot(i_ve['petal.length'],np.zeros_like(i_s['petal.length']),'o')
plt.plot(i_vi['petal.length'],np.zeros_like(i_s['petal.length']),'o')
plt.show()
```

```python
c,b = np.histogram(i_s['petal.length'],bins = 10,density = True)
pdf = c/sum(c)
cdf = np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)
plt.show()
```

```python
c,b = np.histogram(i_s['petal.length'],bins = 10,density = True)
pdf = c/sum(c)
cdf = np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)
c,b = np.histogram(i_vi['petal.length'],bins = 10,density=True)
pdf = c/sum(c)
cdf  = np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)
c,b = np.histogram(i_ve['petal.length'],bins = 10,density = True)
pdf = c/sum(c)
cdf = np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)
plt.show()
```
