```python
import numpy as np
```

```python
t  = np.arange(1.,5.,0.2)
plt.plot(t,t**2,'b--',label='1')
plt.plot(t,t**2.2,'rs',label='2')
plt.plot(t,t**2.5,'g^',label='3')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [2,3,4,5]
plt.plot(x,y,linewidth = 6.0)
plt.show()
```

```python
x1 = [1,2,3,4,5]
y1 = [2,3,4,5,6]
x2  = [3,4,5,6,7]
y2 = [9,8,3,4,5]
plt.plot(x1,y1,x2,y2)
```

```python
t1 = np.arange(0,10,2)
t2 = np.arange(10,20,2)
plt.figure(1)
plt.subplot(211)
plt.grid
plt.plot(t1,np.sin(t1),'b--')
plt.subplot(212)
plt.grid()
plt.plot(t2,np.cos(t1),'g--')
plt.show()
```
